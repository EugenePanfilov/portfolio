# 13. У какого режиссера самые большие суммарные кассовые сбооры?

```python
import pandas as pd
data = pd.read_csv('movie_bd_v5.csv')
directors = pd.Series(0, set(data['director'].str.cat(sep='|').split('|')))
for director in directors.index:
    directors[director] = data[data.director.str.contains(director)].revenue.sum()
directors.sort_values(ascending=False).head(1)
```

# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
data_production_companies = pd.Series(0, set(data['production_companies'].str.cat(sep='|').split('|')))

for company in data_production_companies.index:
    data_production_companies[company] = data[data.production_companies.str.contains(company)].original_title.str.len().mean()
data_production_companies.sort_values(ascending = False)
```
